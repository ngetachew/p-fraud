```python
df_list = []
for item in df.columns:
    dataframe = df.groupby('account_type')[item].describe()
    dataframe['Characteristics'] = item
    df_list.append(dataframe)
```

```python
detailed_stats = pd.concat(df_list, join = 'outer')
detailed_stats = detailed_stats.drop(['25%', '50%', '75%'], axis = 1)
detailed_stats
```

```python
first_column = detailed_stats.pop('Characteristics')
detailed_stats.insert(0, 'Characteristics', first_column)
```
